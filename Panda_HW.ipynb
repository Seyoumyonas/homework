```python
import pandas as pd 
from pathlib import Path
%matplotlib inline 
import datetime as dt
import numpy as np
```

```python
#import the files 
path_tesla=( "../pybank/TeslaHW.csv")
tesla=pd.read_csv(path_tesla, index_col = "Date", infer_datetime_format=True)

path_dsl=("../pybank/DSL.csv")
dsl=pd.read_csv(path_dsl, index_col = "Date", infer_datetime_format=True )

path_app=("../pybank/APPLE.csv")
appl=pd.read_csv(path_app, index_col="Date", infer_datetime_format=True )
```

```python
tesla.plot()
```

```python
appl.plot()
```

```python
dsl.plot()
```

```python
dsl.shape
```

```python
#concat and join the three datas
join_data= pd.concat([tesla,dsl,appl], axis ="columns" , join = "inner" )
join_data.head()
```

```python
join_data.plot(kind="LINE")
```

```python
# Standard deviation of joined data 
join_std = join_data.std()

join_std.head()
```

```python
#sharp ratio  

sharp_ratio =((join_std.mean()*252) - 0.0025)/ (join_std.sum()* np.sqrt(252))
sharp_ratio
```
